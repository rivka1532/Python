```python
years = list(range(2010, 2021))
consumption_oil = [100, 105, 110, 120, 125, 130, 135, 140, 145, 150, 155]
consumption_gas = [80, 82, 85, 88, 90, 95, 100, 105, 110, 115, 120]
consumption_renewable = [20, 25, 30, 35, 40, 45, 50, 60, 70, 80, 90]
```

```python
fig, axs = plt.subplots(1, 3, figsize=(18, 5))
axs[0].plot(years, consumption_oil, marker='o', color='brown')
axs[0].set_title('Oil Consumption')
axs[0].set_xlabel('Year')
axs[0].set_ylabel('Consumption')
axs[1].plot(years, consumption_gas, marker='o', color='blue')
axs[1].set_title('Gas Consumption')
axs[1].set_xlabel('Year')
axs[1].set_ylabel('Consumption')
axs[2].plot(years, consumption_renewable, marker='o', color='green')
axs[2].set_title('Renewable Energy Consumption')
axs[2].set_xlabel('Year')
axs[2].set_ylabel('Consumption')
```
